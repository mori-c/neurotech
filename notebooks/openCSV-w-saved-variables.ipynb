````python
class person:
    'doc inside class'
    
    def __init__(self, name):
        'doc init constructor'
        self.xName = name
        
    def show(self, name01, name02):
        'doc inside show'
        print(self.xName)
        print('sum: ', (name01 + name02))
        
    def __del__(self):
        print('del destructor obj: ', self.xName)
        
p = person('mori')
p.show(2, 3)
print(p.__doc__)
print(p.__init__.__doc__)
print(p.show.__doc__)
````

````python
# __init__ inheritance

class User(object):
    def __init__(self, something):
        print('user init called')
        self.something = something
        
    def method(self):
        return self.something
    
class Others(User):
    def __init__(self, something):
        User.__init__(self, something)
        print('other init called')
        self.something = something
        
    def method(self):
        return self.something
    
newObject = Others('mori')
print(newObject)

# https://www.journaldev.com/18397/python-class-init
````

````python
class rectangle:
    def __init__(self, length, breadth, unitCost = 0):
        self.length = length
        self.breadth = breadth
        self.unitCost = unitCost
        
    def getPerimeter(self):
        return 2 * (self.length + self.breadth)
    
    def getArea(self):
        return self.length * self.breadth
    
    def calculateCost(self):
        area = self.getArea()
        return area * self.unitCost
    
breadth = 120
length = 160
cm2Rrs = 2000

r = rectangle(breadth, length, cm2Rrs)
print('Rectangle Area: %s cm^2' % (r.getArea()))
print('Rectangle Field: Rs. %s ' % (r.calculateCost()))

https://micropyramid.com/blog/understand-self-and-__init__-method-in-python-class/
````
